## Visualizing Earnings Based On College Majors

The purpose of this project is to look at college graduates of different majors between 2010-2012 using different visualizations. The data comes from the American Community Survey.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Importing the data into a dataframe with Pandas
recent_grads = pd.read_csv('recent-grads.csv')

#Displaying first row of dataframe as a table
print(recent_grads.iloc[0])
```

```python
#Looking at first 5 rows and last 5 rows to get a sense of how dataframe is structured
print(recent_grads.head())
print(recent_grads.tail())
```

```python
#Getting summary statistics for all numerical columns in the dataframe
print(recent_grads.describe())
```

## Cleaning up any missing values, so Matplotlib will be able to generate plots

```python
#Determining number of rows
raw_data_count = len(recent_grads.index)
print(raw_data_count)

#Dropping any rows with missing values
recent_grads = recent_grads.dropna()

#Updated row count
cleaned_data_count = len(recent_grads.index)
print(cleaned_data_count)
```

## Generating scatter plots to see if there are any strong visual relationships between different data series

```python
#Scatter plot of Sample_size and Median
recent_grads.plot(x = 'Sample_size', y = 'Median', kind = 'scatter', title = 'Full Time Worker Median Salary vs. Sample Size')

#Determining correlation between Sample_size and Median
corr1 = recent_grads['Sample_size'].corr(recent_grads['Median'])
print('The correlation between Sample_size and Median is {:0.3f}'.format(corr1))
```

```python
#Scatter plot of Sample_size and Unemployment_rate
recent_grads.plot(x = 'Sample_size', y = 'Unemployment_rate', kind = 'scatter', title = 'Unemployment Rate vs. Sample Size')

#Determining correlation between Sample_size and Unemployment_rate
corr2 = recent_grads['Sample_size'].corr(recent_grads['Unemployment_rate'])
print('The correlation between Sample_size and Unemployment_rate is {:0.3f}'.format(corr2))
```

```python
#Scatter plot of Full_time and Median
recent_grads.plot(x = 'Full_time', y = 'Median', kind = 'scatter', title = 'Median Salary vs. Number of Full Time Employees')

#Determining correlation between Full_Time and Median
corr3 = recent_grads['Full_time'].corr(recent_grads['Median'])
print('The correlation between Full_time and Median is {:0.3f}'.format(corr3))
```

```python
#Scatter plot of ShareWomen and Unemployment_rate
recent_grads.plot(x = 'ShareWomen', y = 'Unemployment_rate', kind = 'scatter', title = 'Women vs. Unemployment Rate')

#Determining correlation between ShareWomen and Unemployment_rate
corr4 = recent_grads['ShareWomen'].corr(recent_grads['Unemployment_rate'])
print('The correlation between ShareWomen and Unemployment_rate is {:0.3f}'.format(corr4))
```

```python
#Scatter plot of Men and Median
recent_grads.plot(x = 'Men', y = 'Median', kind = 'scatter', title = 'Full Time Median Salary vs. Men')

#Determining correlation between Men and Median
corr5 = recent_grads['Men'].corr(recent_grads['Median'])
print('The correlation between Men and Median is {:0.3f}'.format(corr5))
```

```python
#Scatter plot of Women and Median
recent_grads.plot(x = 'Women', y = 'Median', kind = 'scatter', title = 'Full Time Median Salary vs. Women')

#Determining correlation between Women and Median
corr6 = recent_grads['Women'].corr(recent_grads['Median'])
print('The correlation between Women and Median is {:0.3f}'.format(corr6))
```

After plotting the different recent_grads series against each other on scatter plots and calculating correlations for each plot, all the correlations were within +/- 0.2 of 0, which suggests there are no relationships between the series plotted.
