```python
#This script will be testing manipulation being done to a .csv file.
import pandas as pd
import numpy as np
import datetime

#Import 311 Data as a dataframe
datafile = '311_data.csv'
data = pd.read_csv(datafile)
data
```

```python
#Create a new data frame with 'Created Date','Closed Date','Borough','Descriptor','Complaint Type', 'Agency', 'Longitude', 'Latitude', 'Status' columns from the 311 Data 
select_data = data[['Created Date','Closed Date','Borough','Descriptor','Complaint Type', 'Agency', 'Longitude', 'Latitude', 'Status']].copy()
select_data
```

```python
#Convert Created Date and Closed Date columns to convert them to datetime objects
select_data['Created Date'] = pd.to_datetime(select_data['Created Date'], format='%m/%d/%Y %I:%M:%S %p')
select_data['Closed Date'] = pd.to_datetime(select_data['Closed Date'], format='%m/%d/%Y %I:%M:%S %p')

#Create a new column called processing time
select_data['processing_time'] = select_data['Closed Date'] - select_data['Created Date']
select_data
```

```python
#Create a new column called start time window
select_data['start_time_window'] = select_data['Created Date'].apply(lambda x: x.hour)
select_data
```

```python
#Ouput new dataframe to a .csv file
select_data.to_csv('311_data(cleaned).csv', index = False)
```
